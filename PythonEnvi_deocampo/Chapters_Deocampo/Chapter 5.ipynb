**Exercise 1**

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

````python
a = 2
b = 2.0
c = a

# A list containing our variables
numbers = [2, 2.0, c, 3]

print(f"Target value (a) is: {a}\n")

for item in numbers:
    # Check for value equality
    equals_check = (item == a)

    # Check for object identity
    is_check = (item is a)

    print(f"Testing item: {item}")
    print(f"  - item == a: {equals_check}")
    print(f"  - item is a: {is_check}")
    print("-" * 20)
````

**Challenge 1**

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333
    
[PEW] E5.1

````python
total = 0
count = 0

while True:
    inp = input("Enter a number: ")

    if inp.lower() == "done":
        break

    try:
        num = float(inp)
    except ValueError:
        print("Invalid input")
        continue

    total += num
    count += 1

if count > 0:
    average = total / count
    print("Sum:", total, "Count:", count, "Average:", average)
else:
    print("No numbers entered")
````

**Challenge 2**

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

[PEW] E5.2

````python
total = 0
count = 0
maximum = None
minimum = None

while True:
    inp = input("Enter a number: ")

    if inp.lower() == "done":
        break

    try:
        num = float(inp)
    except ValueError:
        print("Invalid input")
        continue

    if maximum is None or num > maximum:
        maximum = num

    if minimum is None or num < minimum:
        minimum = num

    count += 1

if count > 0:
    print("Maximum:", maximum)
    print("Minimum:", minimum)
else:
    print("No numbers entered")
````
